```python
import pandas as pd
import matplotlib.pyplot as plt
import keras
keras.__version__
```

```python
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')

train[:5]
```

```python
train.info()  # --> lose data: Age, Cabin, Embarked
```

```python
y_train = train.iloc[:, 1]
X_train = train.drop(['PassengerId', 'Survived', 'Cabin', 'Name', 'Ticket'], axis=1)
```

```python
# Survive vs No Survived
labels = 'no survived', 'survived'
size = y_train.value_counts()
plt.pie(size, labels=labels, autopct=lambda p: '{:.2f}% ({:.0f})'.format(p, (p/100)*size.sum()))
plt.show()
```

```python
print(X_train['Sex'].unique())
X_train['Sex'] = X_train['Sex'].map({ 'male':0, 'female':1 }) 
print(X_train['Embarked'].unique())
X_train['Embarked'] = X_train['Embarked'].map({ 'S':0, 'C':1, 'Q':2 })
```

```python
# 找出空值index
index_nan_emb = X_train[X_train['Embarked']!=X_train['Embarked']].index
index_nan_age = X_train[X_train['Age']!=X_train['Age']].index
index_nan = index_nan_emb.append(index_nan_age)
```

```python
X_train = X_train.drop(index=index_nan)
y_train = y_train.drop(index=index_nan)
```

```python
# onehot-encoder
y_train = keras.utils.to_categorical(y_train, 2)
```

```python
print(X_train.shape)
print(y_train.shape)
```

```python
# use keras Senquential
from keras.models import Sequential
from keras.layers import Dense, Dropout
model = Sequential()
model.add(Dense( units=256, input_dim=7, kernel_initializer='RandomNormal', activation='relu' ))
model.add(Dropout(0.5))
model.add(Dense( units=128, kernel_initializer='RandomNormal', activation='relu' ))
# model.add(Dropout(0.5))
model.add(Dense( units=64, kernel_initializer='RandomNormal', activation='relu' ))
# model.add(Dropout(0.5))


# output layer
model.add(Dense( units=2, kernel_initializer='RandomNormal', activation='sigmoid' ))

# model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# train model
train_history = model.fit(x=X_train.values, y=y_train, validation_split=0.2, epochs=250, batch_size=256, verbose=1)
```

```python
def show_train_history(train_history, train, validation, title, ylabel):
    plt.plot(train_history.history[train])
    plt.plot(train_history.history[validation])
    plt.title(title)
    plt.ylabel(ylabel)
    plt.xlabel('Epoch')
    plt.legend(['train', 'validation'])
    plt.show()
```

```python
# acc
show_train_history(train_history, 'acc', 'val_acc', title='train_history_acc', ylabel='acc')
# loss
show_train_history(train_history, 'loss', 'val_loss', title='train_history_loss', ylabel='loss')
```

```python
model.save('titanic_DNN.h5')
```
